```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
dataset = load_iris()

df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['target'] = dataset.target
df.head()
```

# Exploratory Data Analysis
## Univariate Analysis

```python
sns.countplot(df['target'])
```

## Bivariate Analysis

```python
sns.pairplot(data=df, hue='target')
```

```python
df.boxplot(by="target", figsize=(18, 9))
```

```python
plt.figure(figsize=(10,5))
from pandas.plotting import parallel_coordinates
parallel_coordinates(df,'target')
```

* For an instance to belong to class 0, it should have lower sepal length, petal length and width but a higher sepal width. 
* Given only petal length and width, an instance can be identfied if it belongs to class 0 or not.
* An instnace belonging to class 2, has the highest sepal length, petal length and width.

```python
X = df.drop('target',1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

from sklearn.neighbors import KNeighborsClassifier 
nb = KNeighborsClassifier(n_neighbors=3)

nb.fit(X,y)
pred = nb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
