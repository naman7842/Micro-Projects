# Data Cleaning 

#### Null values in HDI for year

```python
df['HDI for year'].isnull().sum()
```

```python
c_1 = df["HDI for year"]
c_2 = df["suicides_no"]
corr = c_1.corr(c_2)
corr.round(2)
```

Since, HDI and no. of suicides are not correlated much, so dropping the feature for now.

```python
df.drop('HDI for year', inplace=True, axis=1)
```

#### gdp_for_year object to int

```python
df[' gdp_for_year ($) '] = df[' gdp_for_year ($) '].str.replace(',','')
```

```python
df[' gdp_for_year ($) '].astype(float)
```

#### Extract min, max, avg age from age feature.

```python
print(df['age'].str.find("+ years").value_counts())
print(df['age'].str.find("75+ years").value_counts())
```

* Since age feature containes two kinds of values, one range and the other 75+years, replacing them with "- 75" so average age will be 75 for the latter.

```python
df['age'] = df['age'].apply(lambda x: x.replace("+ years", "-75 years" ))
df['age'] = df['age'].apply(lambda x: x.replace(" years", "" ))
```

```python
df['min_age'] = df['age'].apply(lambda x: x.split('-')[0]).astype(int)
df['max_age'] = df['age'].apply(lambda x: x.split('-')[1]).astype(int)
```

```python
df['avg_age'] = (df['min_age']+df['max_age'])/2
```

## Q1. Which country has the maximum number of suicides over the given period?

```python
df_by_country = df.groupby(['country']).suicides_no.sum()
df_by_country = pd.DataFrame(df_by_country)
df_by_country = df_by_country.sort_values(by = 'suicides_no', ascending = False)
df_by_country.head()
```

The countries with maximum number of suicides are as follows :
* Russian Federation	
* United States
* Japan
* France
* Ukraine

```python
df_by_country.plot(kind='barh', figsize=(5,19))
```

## Q6. How does the gdp per capita of a country for a particular year affect the number of suicices?

```python
df1 = pd.pivot_table(df, index = ['country', 'year'],
                     values=['suicides_no','gdp_per_capita ($)', 'population', 'suicides/100k pop'],
                     aggfunc = {"suicides_no" : np.sum
                               ,"gdp_per_capita ($)" : np.mean
                               ,"population" : np.mean
                               ,"suicides/100k pop" : np.mean})
```
